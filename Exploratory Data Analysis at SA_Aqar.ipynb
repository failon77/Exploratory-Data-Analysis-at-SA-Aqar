 # 1. Importing the required libraries for EDA


```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

# 2. Loading the data into the data frame.


```python
df = pd.read_csv("data/SA_Aqar.csv")
```

```python
df.head()
```

```python
df.tail(5)  
```

# 3. Checking the types of data

```python
df.dtypes
```

# 4. Dropping irrelevant columns

```python
df = df.drop(['roof', 'pool', 'frontyard', 'basement', 'stairs', 'elevator', 'fireplace','driver_room','maid_room','furnished','ac'], axis=1)
df.head(5)
```

# 5. Dropping the duplicate rows

```python
 df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

# 6. Dropping the missing or null values.

```python
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())   # After dropping the values
```

# 7. Detecting Outliers

```python
sns.boxplot(x=df['property_age'])
```

```python
df.city.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of district by city")
plt.ylabel('Number of district')
plt.xlabel('city');
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['livingrooms'], df['price'])
ax.set_xlabel('livingrooms')
ax.set_ylabel('price')
plt.show()
```
